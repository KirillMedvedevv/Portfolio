```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d, griddata
import imageio
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 1 Чтение большого CSV файла
def read_large_csv(file_path, chunksize=1000):
    """Читает большой CSV файл по частям и возвращает объединённый DataFrame."""
    chunks = []
    for chunk in pd.read_csv(file_path, sep=';', chunksize=chunksize):
        chunks.append(chunk)
    return pd.concat(chunks, ignore_index=True)

# 2 Визуализация данных для конкретного ti

def visualize_for_ti(df1, df2, ti, output_dir='plots', m=14, c=1, L=1):
    """Визуализирует данные u(x) для заданного ti для двух наборов данных, с интерполяцией при необходимости."""
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    # Функция для получения данных для заданного ti
    def get_data_for_ti(df, ti):
        if ti in df['t'].unique():
            return df[df['t'] == ti]
        else:
            x_unique = df['x'].unique()
            u_interp = []
            for x in x_unique:
                data_x = df[df['x'] == x]
                if len(data_x) > 1:
                    f = interp1d(data_x['t'], data_x['u'], kind='cubic', fill_value="extrapolate")
                    u_interp.append(f(ti))
                else:
                    u_interp.append(np.nan)
            return pd.DataFrame({'x': x_unique, 'u': u_interp})

    # Получаем данные для обоих наборов
    data1 = get_data_for_ti(df1, ti)
    data2 = get_data_for_ti(df2, ti)


    plt.figure()
    plt.plot(data1['x'], data1['u'], 'b-', label='Implicit')
    plt.plot(data2['x'], data2['u'], 'r--', label='Explicit')
    #plt.plot(data1['x'], np.sin(np.pi*m*(data1['x'] - c*ti)/L),'g-', label='Analytic')
    plt.ylim(1, 3)
    plt.grid()
    plt.xlabel(r'$x$')
    plt.ylabel('u')
    plt.title(f'CFD Data at t = {ti}')
    plt.legend()
    plot_path = os.path.join(output_dir, f'plot_t_{ti}.png')
    plt.savefig(plot_path, dpi=300)
    plt.close()
    return plot_path

# 3. Построение GIF анимации
def create_gif(df1, df2, tau, num_frames=100, output_gif='s_num.gif'):
    """Создаёт GIF анимацию для интервала [0, tau] для двух наборов данных."""
    times = np.linspace(0, tau, num_frames)
    images = []
    for ti in times:
        plot_path = visualize_for_ti(df1, df2, ti)
        images.append(imageio.imread(plot_path))

    imageio.mimsave(output_gif, images, duration=0.05)  # Сохраняем GIF с длительностью кадра 0.1 сек

def auto_build(df1, df2):
    fig, ax = plt.subplots(figsize=(10, 6))

    # Построение точечного графика u от eta
    #ax.plot(df1['eta'], df1['u'], 'b-', label='Analytic')
    ax.scatter(df2['eta'], df2['u'], s=1, label='Numeric')

    # Настройка подписей осей и заголовка
    ax.set_xlabel(r'$\eta = \frac{x}{\sqrt{t}}$', fontsize=14)
    ax.set_ylabel(r'$u$', fontsize=14)
    ax.set_title('Зависимость скорости в автомодельных координатах', fontsize=16)

    # Добавление major и minor gridlines
    ax.grid(which='major', linestyle='-', linewidth=0.5, color='black')
    ax.grid(which='minor', linestyle=':', linewidth=0.5, color='gray')
    ax.minorticks_on()

    # Отображение графика
    fig.savefig('Auto.png')
    fig.show()


def get_interpolated_value(df, x_target, t_target, method='linear'):
    # Проверка точного совпадения
    exact_match = df[(df['x'] == x_target) & (df['t'] == t_target)]
    if not exact_match.empty:
        return exact_match['u'].values[0], 'NORM'

    # Подготовка данных для интерполяции
    points = df[['x', 't']].values
    values = df['u'].values
    target = np.array([[x_target, t_target]])

    # Интерполяция выбранным методом
    interpolated = griddata(points, values, target, method=method)

    # Если результат NaN, пробуем метод nearest
    if np.isnan(interpolated):
        interpolated = griddata(points, values, target, method='nearest')

    return interpolated[0]
if __name__ == "__main__":
    file_path1 = 'sin_5_EXP.csv' #IMPLICIT
    file_path2 = 'sin_5_EXP.csv' #EXPLICIT
    #df1 = read_large_csv(file_path1)
    #df2 = read_large_csv(file_path2)

    #visualize_for_ti(df1, df2, 0, output_dir='my', m=5, L=1)
    #visualize_for_ti(df1, df2, 5, output_dir='my', m=5, L=1)

    #file_path1 = 'sin_14_IMP.csv' #IMPLICIT
    #file_path2 = 'sin_14_EXP.csv' #EXPLICIT
    df1 = read_large_csv(file_path1)
    df2 = read_large_csv(file_path2)

    #visualize_for_ti(df1, df2, 0, output_dir='my', m=5, L=1)
    visualize_for_ti(df1, df2, 1, output_dir='my', m=5, L=1)
    #visualize_for_ti(df1, df2, 53, output_dir='my')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Параметры сетки
sigma = np.linspace(0, 1, 100)        # Число Куранта
theta = np.linspace(0, np.pi, 100)   # Безразмерное волновое число
sigma_grid, theta_grid = np.meshgrid(sigma, theta)

# Вычисление модуля коэффициента перехода
def calculate_G(scheme_type):
    if scheme_type == "explicit":
        return np.sqrt(1 + 2*sigma_grid*(sigma_grid-1)*(1-np.cos(theta_grid)))
    elif scheme_type == "implicit":
        return 1 / np.sqrt(1 + 2*sigma_grid + 2*sigma_grid**2 - 2*sigma_grid*(1 + sigma_grid)*np.cos(theta_grid))

G_explicit = calculate_G("explicit")
G_implicit = calculate_G("implicit")

# Настройка графики
plt.rcParams.update({'font.size': 12})
fig = plt.figure(figsize=(14, 6))

# Явная схема
ax1 = fig.add_subplot(121, projection='3d')
surf1 = ax1.plot_surface(
    sigma_grid,
    theta_grid,
    G_explicit,
    cmap='coolwarm',
    rstride=2,
    cstride=2,
    alpha=0.8
)
ax1.set_title('A) Явная противопоточная схема', pad=15)
ax1.set_xlabel('σ (Курант)', labelpad=10)
ax1.set_ylabel('θ (волновое число)', labelpad=10)
ax1.set_zlabel('|G|', labelpad=5)
ax1.view_init(elev=25, azim=-45)

# Неявная схема
ax2 = fig.add_subplot(122, projection='3d')
surf2 = ax2.plot_surface(
    sigma_grid,
    theta_grid,
    G_implicit,
    cmap='viridis',
    rstride=2,
    cstride=2,
    alpha=0.8
)
ax2.set_title('B) Неявная противопоточная схема', pad=15)
ax2.set_xlabel('σ (Курант)', labelpad=10)
ax2.set_ylabel('θ (волновое число)', labelpad=10)
ax2.set_zlabel('|G|', labelpad=5)
ax2.view_init(elev=25, azim=-45)

# Общие настройки
plt.tight_layout()
fig.colorbar(surf1, ax=ax1, shrink=0.5, pad=0.1)
fig.colorbar(surf2, ax=ax2, shrink=0.5, pad=0.1)
plt.tight_layout()
plt.savefig('Dissipation.png', dpi=300)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d

def plot_convection_solutions(csv_file, t_target, k_wave=1, m=1, c=1.0, h=0.01, tau=0.01, save_path='convection_plot.png', typ='implicit'):
    """
    Строит график аналитического решения, оценки численного решения и интерполированного численного решения
    для уравнения конвекции с начальным условием sin(pi * k * x), периодичным на [0, L].

    Parameters:
    - csv_file: путь к CSV-файлу с данными (формат: t, x, u)
    - t_target: момент времени для построения графика
    - k_wave: волновое число k в sin(pi * k * x)
    - m: число полных волн на отрезке [0, L]
    - c: скорость переноса
    - h: шаг по пространству
    - tau: шаг по времени
    - save_path: путь для сохранения графика
    """
    # Вычисление длины области L для периодичности
    L = 2 * m / k_wave  # L = 2m / k для sin(pi * k * x)
    sigma = c * tau / h  # число Куранта
    k = np.pi * k_wave  # волновое число для sin(pi * k * x)
    n_steps = int(t_target / tau)  # число временных шагов

    # Чтение CSV-файла
    data = pd.read_csv(csv_file, sep=';')
    data.columns = ['t', 'x', 'u']  # предполагаем формат t, x, u

    # Фильтрация данных для ближайших временных слоев
    times = data['t'].unique()
    t_before = times[times <= t_target][-1] if any(times <= t_target) else times[0]
    t_after = times[times >= t_target][0] if any(times >= t_target) else times[-1]

    # Данные для интерполяции
    data_before = data[data['t'] == t_before][['x', 'u']].sort_values('x')
    data_after = data[data['t'] == t_after][['x', 'u']].sort_values('x')

    # Пространственная сетка
    x_common = np.linspace(0, L, 1000)  # единая сетка для всех решений

    # Интерполяция численного решения
    if t_before == t_after:
        u_numerical = interp1d(data_before['x'], data_before['u'], kind='linear', fill_value="extrapolate")(x_common)
    else:
        u_before = interp1d(data_before['x'], data_before['u'], kind='linear', fill_value="extrapolate")(x_common)
        u_after = interp1d(data_after['x'], data_after['u'], kind='linear', fill_value="extrapolate")(x_common)
        alpha = (t_target - t_before) / (t_after - t_before)
        u_numerical = (1 - alpha) * u_before + alpha * u_after

    # Аналитическое решение: u(x,t) = sin(pi * k * (x - c * t))
    def analytical_solution(x, t, c, k):
        return np.sin(k * (x - c * t))

    # Вычисление коэффициента усиления xi
    def compute_xi(sigma, k, h, type='explicit'):
        if type=='explicit':
            xi = 1 - sigma*(1-np.exp(1j*h*k))
        else:
            xi = 1/(1+sigma*(1 - np.exp(1j*k*h)))
        return xi

    # Модуль xi
    def compute_abs_xi(xi):
        return np.abs(xi)

    # Числовая фазовая скорость
    def compute_c_num(xi, k, tau):
        phi = np.angle(xi)
        return phi / (k * tau)

    # Оценка численного решения
    def estimated_numerical_solution(x, t, xi, c_num, k, n_steps):
        abs_xi = compute_abs_xi(xi)
        amplitude = abs_xi ** n_steps
        phase = k * (x - c_num * t)
        return amplitude * np.sin(phase)

    # Вычисление xi и c_num
    xi = compute_xi(sigma, k, h, typ)
    c_num = compute_c_num(xi, k, tau)

    # Вычисление решений
    u_analytical = analytical_solution(x_common, t_target, c, k)
    u_estimated = estimated_numerical_solution(x_common, t_target, xi, c_num, k, n_steps)

    # Построение графика
    plt.figure(figsize=(10, 6))
    plt.plot(x_common, u_analytical, label='Аналитическое решение', linewidth=2, color='blue')
    plt.plot(x_common, u_estimated, label='Оценка численного решения', linestyle='--', linewidth=2, color='red')
    plt.plot(x_common, u_numerical, label='Численное решение', linestyle='-.', linewidth=2, color='green')

    # Настройка осей
    plt.xlabel('x')
    plt.ylabel('u')
    plt.title(f'Сравнение решений для t={t_target:.2f}, k={k_wave}, L={L:.2f}, σ={sigma:.2f}')

    # Настройка сетки
    plt.grid(True, which="major", linestyle='-', linewidth=0.5)
    plt.grid(True, which="minor", linestyle=':', linewidth=0.2)
    plt.minorticks_on()

    plt.legend()
    plt.tight_layout()
    plt.savefig(save_path)
    plt.show()
    plt.close()

# Пример использования
plot_convection_solutions(
    csv_file='sin_14_EXP.csv',
    t_target=0.2,
    k_wave=14,  # нечетная гармоника
    m=7,       # одна полная волна
    c=1.0,
    h=1/200,
    tau=1/400,
    save_path='14_explicit.png',
    typ='explicit'
)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Шаг 1: Импорт данных
data = pd.read_csv('sin_5_IMP.csv', sep=';')

# Шаг 2: Параметры задачи
m = 5  # волновое число
c = 1  # скорость волны
L = 2 / m  # длина области
delta_t = L/400  # шаг по времени
delta_x = L/200  # шаг по пространству
sigma = c * delta_t / delta_x
k = np.pi * m  # волновое число

# Шаг 3: Аналитическое решение
def analytical_solution(x, t, m, c):
    return np.sin(np.pi * m * (x - c * t))

# Шаг 4: Оценка аналитического решения с учетом схемы
def analytical_solution_with_estimate(x, t, m, c, delta_t, delta_x, sigma, k, scheme_type):
    if scheme_type == 'explicit':
        # Для явной схемы
        cos_term = np.cos(k * delta_x)
        sin_term = np.sin(k * delta_x)
        real_part = 1 - sigma * (1 - cos_term)
        imag_part = sigma * sin_term
        lambda_modulus = np.sqrt(real_part**2 + imag_part**2)
        lambda_arg = -np.arctan2(imag_part, real_part)
        alpha = -np.log(lambda_modulus) / delta_t
        c_disp = lambda_arg / (k * delta_t)
    elif scheme_type == 'implicit':
        # Для неявной схемы
        g = 1/np.sqrt(1+sigma*(1 - np.exp(1j*k*delta_x)))
        #sin_term = np.sin(k * delta_x)
        print(np.imag(g))
        lambda_modulus = np.sqrt(np.real(g)**2 + np.imag(g)**2)
        lambda_arg = -np.arctan2(np.imag(g), np.real(g))
        alpha = -np.log(lambda_modulus) / delta_t
        c_disp = -lambda_arg / (k * delta_t)
    else:
        raise ValueError("Неизвестный тип схемы. Используйте 'explicit' или 'implicit'.")

    return np.exp(-alpha * t) * np.sin(np.pi * m * (x - c_disp * t))

# Шаг 5: Построение графиков
def plot_solution(t_selected, scheme_type, sigma):
    data_t = data[data['t'] == t_selected]
    x_analytical = np.linspace(0, L, 1000)
    u_analytical = analytical_solution(x_analytical, t_selected, m, c)
    u_analytical_estimate = analytical_solution_with_estimate(x_analytical, t_selected, m, c, delta_t, delta_x, sigma, k, scheme_type)

    plt.figure(figsize=(10, 6))
    plt.plot(x_analytical, u_analytical, label='Аналитическое решение', color='blue')
    plt.plot(data_t['x'], data_t['u'], label='Численное решение', color='red')
    plt.plot(x_analytical, u_analytical_estimate, label='Аналитическое с оценкой', color='green', linestyle='--')

    plt.grid(which='major', linestyle='-', linewidth=0.5, color='black')
    plt.grid(which='minor', linestyle=':', linewidth=0.5, color='gray')
    plt.minorticks_on()

    plt.legend()
    plt.xlabel('x')
    plt.ylabel('u')
    plt.title(f'Решения при t = {t_selected} ({scheme_type})')
    #plt.savefig(f'solutions_t_{t_selected}_{scheme_type}.png')

# Пример вызова
t_selected = 4
#plot_solution(t_selected, 'explicit')
plot_solution(t_selected, 'implicit', sigma)
```
